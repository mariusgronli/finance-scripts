```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bam_price = yf.download(tickers="CSCO",start="2017-01-01",end="2019-12-11",interval="1d")
```

```python
bam_price.drop(["Open","High","Low","Close","Volume"],axis=1,inplace=True)
```

```python
bam_ret = bam_price.pct_change(1)[1:]
```

```python
bam_ret.columns=["returns"]
```

```python
bam_ret.plot()
```

```python
%matplotlib notebook
bam_price['Close: 30 Day Mean'] = bam_price['Adj Close'].rolling(window=60).mean()
bam_price['Upper'] = bam_price['Close: 30 Day Mean'] + 2*bam_price['Adj Close'].rolling(window=60).std()
bam_price['Lower'] = bam_price['Close: 30 Day Mean'] - 2*bam_price['Adj Close'].rolling(window=60).std()
bam_price[['Adj Close','Close: 30 Day Mean','Upper','Lower']].plot(figsize=(12,6))
```
